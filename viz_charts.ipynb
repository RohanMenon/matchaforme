```python
rate = ratings['rating'].astype('category')
rate_count = rate.groupby(rate).apply(lambda x: x.count())
```

```python
sns.barplot(x=rate_count.index, y=rate_count.values, palette='viridis')
plt.yticks(range(0, int(rate_count.max()*2)+1, 2))
plt.xlabel('Rating')
plt.ylabel('Rating Count')
plt.title('Distribution of Ratings')
```
